## Holiday and non Holiday
### I want to know the willingness of people renting bike during holiday and non holiday (value 0 is non holiday and value 1 is holiday). I will find out by comparing the count (cnt) of people renting bike during holiday and non holiday. Also, I will calculate the max, min and average value for both holiday and non holiday.

```python
df_holiday_1 = df[df['holiday']==1]
df_holiday_1
```

```python
df_holiday_0 = df[df['holiday']==0]
df_holiday_0
```

```python
sns.boxplot(x='holiday', y='cnt', data = df)
```

```python
column0 = df_holiday_0['cnt']
max_value_not_holiday = column0.max()
column1 = df_holiday_1['cnt']
max_value_holiday = column1.max()
print('Max count not holiday:', max_value_not_holiday, 'Max count holiday:', max_value_holiday)
min_value_not_holiday = column0.min()
min_value_holiday = column1.min()
print('Min count not holiday:', min_value_not_holiday, 'Min count holiday:', min_value_holiday)
ave_value_not_holiday = column0.mean()
ave_value_holiday = column1.mean()
print('Ave count not holiday:', ave_value_not_holiday, 'Ave count holiday', ave_value_holiday)
```

### As the result shows, we can see that surprisingly people rented more bike on average compare to holidays (4527:3735). Holidays have a more steady rent rate compare to non holidays, ranging from 1000 counts to 7403 counts compare to 22 counts to 8714 counts.

## Temperature and count
### I want to see how the temperature affect people's willingness to rent a bike. Temperature value is derived by (t-t_min)/(t_max-t_min), t_min=-8, t_max=+39 (only in hourly scale). This is the default value for temperature, but we can still see whether people are more willing to rent a bike during hot days or cooler days.

```python
df_temp = df[['temp', 'cnt']]
df_temp
```

```python
sns.kdeplot(x = 'temp', data = df_temp)
```

```python
sns.boxplot(x='temp', data=df_temp)
```

### The boxplot also shows similar result with the kdeplot. Temperature values mostly falls between 0.32 - 0.65.

```python
sns.lineplot(x='temp',y='cnt',data=df_temp)
```

### Through lineplot we can clearly see that people prefer biking on a hotter day compare to cooler day, which is another surprice. My initial hypothesis is the people will prefer cooler day than hotter day to enjoy the breeze. We can assume that data is collected in a colder area.

## Humidity and count

```python
df_hum = df[['hum','cnt']]
df_hum
```

```python
sns.boxplot(x='hum',data=df_hum)
```

### After the previous result from temperature, I believe one of boxplot or kdeplot is enough to show the density of the value we want to see.
